```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import time
from sklearn.datasets import make_moons, make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

### Neural Network Class

```python
class ANN:
    def __init__(self, num_layers, npl):
        self.num_layers = num_layers
        self.npl = npl
        self.params = {}
        self.costs = []
        self.accuracy = []
        self.val_costs = []
        self.val_accuracy = []
        
        for l in range(1, self.num_layers):
            self.params[f'w{l}'] = np.random.randn(npl[l], npl[l - 1]) * np.sqrt(2 / npl[l - 1]) ### He Weight Initialization for ReLU
            self.params[f'b{l}'] = np.zeros((npl[l], 1))

    def relu(self, z):
        z = np.array(z)
        return np.maximum(0, z)

    def sigmoid(self, z):
        z = np.array(z)
        return 1 / (1 + np.exp(-z))

    def forward_prop(self, x):
        fwd_outputs = {'a0': x}
        a = x

        for l in range(1, self.num_layers - 1):
            w = self.params[f'w{l}']
            b = self.params[f'b{l}']
            z = np.dot(w, a) + b
            a = self.relu(z)
            fwd_outputs[f'z{l}'] = z
            fwd_outputs[f'a{l}'] = a

        final = self.num_layers - 1
        w = self.params[f'w{final}']
        b = self.params[f'b{final}']
        z = np.dot(w, a) + b
        a_final = self.sigmoid(z)
        fwd_outputs[f'z{final}'] = z
        fwd_outputs[f'a{final}'] = a_final

        return a_final, fwd_outputs

    def find_cost(self, a_final, y):
        n = y.shape[1]
        cost = - (1 / n) * np.sum(y * np.log(a_final + 1e-8) + (1 - y) * np.log(1 - a_final + 1e-8)) ### 1e-8 is added to prevent log(0)
        return np.squeeze(cost)

    def back_prop(self, a_final, y, fwd_outputs):
        gradients = {}
        n = y.shape[1]
        final = self.num_layers - 1

        dz = a_final - y
        a_prev = fwd_outputs[f'a{final - 1}']
        gradients[f'dw{final}'] = (1 / n) * np.dot(dz, a_prev.T)
        gradients[f'db{final}'] = (1 / n) * np.sum(dz, axis=1, keepdims=True)

        for l in reversed(range(1, self.num_layers - 1)):
            da = np.dot(self.params[f'w{l + 1}'].T, dz)
            a = fwd_outputs[f'a{l}']
            dz = da * (a > 0).astype(float)
            a_prev = fwd_outputs[f'a{l - 1}']
            gradients[f'dw{l}'] = (1 / n) * np.dot(dz, a_prev.T)
            gradients[f'db{l}'] = (1 / n) * np.sum(dz, axis=1, keepdims=True)

        return gradients

    def gradient_descent(self, gradients, alpha):
        for l in range(1, self.num_layers):
            self.params[f'w{l}'] -= alpha * gradients[f'dw{l}']
            self.params[f'b{l}'] -= alpha * gradients[f'db{l}']

    def training(self, x_train, y_train, x_val, y_val, epochs, batch_size, alpha):
        x_train_t = x_train.T
        num_classes = self.npl[-1]
        n = x_train.shape[0]

        y_val_t = y_val.reshape(1, -1)

        start = time.time()

        for i in range(epochs):
            
            epoch_start = time.time()
            epoch_cost = 0
            
            for j in range(0, n, batch_size): ### Mini Batch Gradient Descent
                X_batch = x_train_t[:, j:j + batch_size]
                y_batch = y_train[j:j + batch_size]
                y_batch_t = y_batch.reshape(1, -1)

                A_final, forward_outputs = self.forward_prop(X_batch)
                cost = self.find_cost(A_final, y_batch_t)
                epoch_cost += cost

                grads = self.back_prop(A_final, y_batch_t, forward_outputs)
                self.gradient_descent(grads, alpha)

            A_final_val, forward_outputs_val = self.forward_prop(x_val.T)
            cost_val = self.find_cost(A_final_val, y_val_t)
            self.val_costs.append(cost_val)

            avg_epoch_cost = epoch_cost / (n / batch_size)
            self.costs.append(avg_epoch_cost)

            epoch_accuracy = accuracy_score(y_train, self.predict(x_train_t))
            self.accuracy.append(epoch_accuracy)
            val_accuracy = accuracy_score(y_val, self.predict(x_val.T))
            self.val_accuracy.append(val_accuracy)

            epoch_end = time.time()
            print(f"Epoch {i + 1}/{epochs} - Cost: {avg_epoch_cost:.4f} Val_Cost: {cost_val:.4f} Accuracy: {epoch_accuracy:.4f} Val_Accuracy: {val_accuracy:.4f} Time: {epoch_end - epoch_start:.2f}s")

        end = time.time()
        print(f"Training Time: {end - start:.2f}s\n")

    def predict(self, X):
        A_final, _ = self.forward_prop(X)
        return (A_final >= 0.5).flatten() ### returns 1 or 0
```

### Loading the make_classification Dataset

```python
Xl, yl = make_classification(
    n_samples=10000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    n_classes=2,
    class_sep=2.0,
    random_state=42
)
```

```python
X1, X_test1, y1, y_test1 = train_test_split(Xl, yl, test_size=0.2, random_state=42)
X_train1, X_val1, y_train1, y_val1 = train_test_split(X1, y1, test_size=1/8, random_state=42)
```

```python
print(f"Training data shape: {X_train1.shape}")
print(f"Training labels shape: {y_train1.shape}")
print(f"Validation data shape: {X_val1.shape}")
print(f"Validation labels shape: {y_val1.shape}")
print(f"Test data shape: {X_test1.shape}")
print(f"Test labels shape: {y_test1.shape}")
```

### make_classificiation Scatter Plot

```python
plt.scatter(X_train1[:, 0], X_train1[:, 1], c=y_train1)
plt.title("make_classification")
plt.show()
```

### Distribution of Class Labels in make_classification

```python
sns.countplot(x=y_train1, hue=y_train1, legend=False)
plt.show()
```

### Training Model 1: Single Layer Perceptron on Linearly Separable Dataset

```python
INPUT_DIM = X_train1.shape[1]
OUTPUT_DIM = len(np.unique(y_train1))
EPOCHS = 100
ALPHA = 0.05
BATCH_SIZE = 500
layers = [INPUT_DIM, 1]
model_1 = ANN(len(layers),layers)
model_1.training(X_train1, y_train1, X_val1, y_val1, epochs=EPOCHS, batch_size=BATCH_SIZE, alpha=ALPHA)
```

### Model 1 Performance

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(np.array(model_1.costs), label='Training Cost')
axes[0].plot(np.array(model_1.val_costs), label='Validation Cost')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Cost')
axes[0].set_title('Training Cost')
axes[0].legend()

axes[1].plot(np.array(model_1.accuracy), label='Training Accuracy')
axes[1].plot(np.array(model_1.val_accuracy), label='Validation Accuracy')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training Accuracy')
axes[1].legend()
plt.legend()

plt.tight_layout()
plt.show()
```

### Model 1 Predictions

```python
y_pred1 = model_1.predict(X_test1.T)
cm = confusion_matrix(y_test1, y_pred1,)
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Model 1 Evaluation

```python
acc1 = accuracy_score(y_test1, y_pred1)
prec1 = precision_score(y_test1, y_pred1,)
recall1 = recall_score(y_test1, y_pred1,)
f11 = f1_score(y_test1, y_pred1,)
np.set_printoptions(precision=3)

print(f"Accuracy: {acc1:.3f}")
print(f"Precision: {prec1:.3f}")
print(f"Recall: {recall1:.3f}")
print(f"F1 Score: {f11:.3f}")
```
