```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import time
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

### Loading the MNIST Dataset

```python
(X, y), (X_test, y_test) = mnist.load_data()
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=1/6, random_state=42)
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
X_val  = X_val.reshape(X_val.shape[0], -1)
X_test  = X_test.reshape(X_test.shape[0], -1)
```

```python
print(f"Training data shape: {X_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Validation data shape: {X_val.shape}")
print(f"Validation labels shape: {y_val.shape}")
print(f"Test data shape: {X_test.shape}")
print(f"Test labels shape: {y_test.shape}")
```

### Count Plot of Each Digit

```python
sns.countplot(x=y_train, hue=y_train, legend=False)
plt.show()
```

### Normalizing the Pixel Values

```python
X_train = X_train.astype("float32") / 255.0
X_val = X_val.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
```

### Neural Network Class

```python
class ANN:
    def __init__(self, num_layers, npl):
        self.num_layers = num_layers
        self.npl = npl
        self.params = {}
        self.costs = []
        self.accuracy = []
        self.val_costs = []
        self.val_accuracy = []
        
        for l in range(1, self.num_layers):
            self.params[f'w{l}'] = np.random.randn(npl[l], npl[l - 1]) * np.sqrt(1 / npl[l - 1]) ### Xavier Weight Initialization
            self.params[f'b{l}'] = np.zeros((npl[l], 1))

    def sigmoid(self, z):
        z = np.array(z)
        return 1 / (1 + np.exp(-z))

    def softmax(self, z):
        ez = np.exp(z - np.max(z, axis=0, keepdims=True))
        return ez / np.sum(ez, axis=0, keepdims=True)

    def one_hot_encoding(self, y, num_classes):
        n = y.shape[0]
        y_1hot = np.zeros((n, num_classes))
        y_1hot[np.arange(n), y] = 1
        return y_1hot.T

    def forward_prop(self, x):
        fwd_outputs = {'a0': x}
        a = x

        for l in range(1, self.num_layers - 1):
            w = self.params[f'w{l}']
            b = self.params[f'b{l}']
            z = np.dot(w, a) + b
            a = self.sigmoid(z)
            fwd_outputs[f'z{l}'] = z
            fwd_outputs[f'a{l}'] = a

        final = self.num_layers - 1
        w = self.params[f'w{final}']
        b = self.params[f'b{final}']
        z = np.dot(w, a) + b
        a_final = self.softmax(z)
        fwd_outputs[f'z{final}'] = z
        fwd_outputs[f'a{final}'] = a_final

        return a_final, fwd_outputs

    def find_cost(self, a_final, y):
        n = y.shape[1]
        cost = - (1 / n) * np.sum(y * np.log(a_final + 1e-8)) ### 1e-8 is added to prevent log(0)
        return np.squeeze(cost)

    def back_prop(self, a_final, y, fwd_outputs):
        gradients = {}
        n = y.shape[1]
        final = self.num_layers - 1

        dz = a_final - y
        a_prev = fwd_outputs[f'a{final - 1}']
        gradients[f'dw{final}'] = (1 / n) * np.dot(dz, a_prev.T)
        gradients[f'db{final}'] = (1 / n) * np.sum(dz, axis=1, keepdims=True)

        for l in reversed(range(1, self.num_layers - 1)):
            da = np.dot(self.params[f'w{l + 1}'].T, dz)
            a = fwd_outputs[f'a{l}']
            dz = da * a * (1 - a)
            a_prev = fwd_outputs[f'a{l - 1}']
            gradients[f'dw{l}'] = (1 / n) * np.dot(dz, a_prev.T)
            gradients[f'db{l}'] = (1 / n) * np.sum(dz, axis=1, keepdims=True)

        return gradients

    def gradient_descent(self, gradients, alpha):
        for l in range(1, self.num_layers):
            self.params[f'w{l}'] -= alpha * gradients[f'dw{l}']
            self.params[f'b{l}'] -= alpha * gradients[f'db{l}']

    def training(self, x_train, y_train, x_val, y_val, epochs, batch_size, alpha):
        x_train_t = x_train.T
        num_classes = self.npl[-1]
        n = x_train.shape[0]

        y_val_one_hot = self.one_hot_encoding(y_val, num_classes)

        start = time.time()

        for i in range(epochs):
            
            epoch_start = time.time()
            epoch_cost = 0
            
            for j in range(0, n, batch_size): ### Mini Batch Gradient Descent
                X_batch = x_train_t[:, j:j + batch_size]
                y_batch = y_train[j:j + batch_size]
                Y_batch_one_hot = self.one_hot_encoding(y_batch, num_classes)

                A_final, forward_outputs = self.forward_prop(X_batch)
                cost = self.find_cost(A_final, Y_batch_one_hot)
                epoch_cost += cost

                grads = self.back_prop(A_final, Y_batch_one_hot, forward_outputs)
                self.gradient_descent(grads, alpha)

            A_final_val, forward_outputs_val = self.forward_prop(x_val.T)
            cost_val = self.find_cost(A_final_val, y_val_one_hot)
            self.val_costs.append(cost_val)

            avg_epoch_cost = epoch_cost / (n / batch_size)
            self.costs.append(avg_epoch_cost)

            epoch_accuracy = accuracy_score(y_train, self.predict(x_train_t))
            self.accuracy.append(epoch_accuracy)
            val_accuracy = accuracy_score(y_val, self.predict(x_val.T))
            self.val_accuracy.append(val_accuracy)

            epoch_end = time.time()
            print(f"Epoch {i + 1}/{epochs} - Cost: {avg_epoch_cost:.4f} Val_Cost: {cost_val:.4f} Accuracy: {epoch_accuracy:.4f} Val_Accuracy: {val_accuracy:.4f} Time: {epoch_end - epoch_start:.2f}s")

        end = time.time()
        print(f"Training Time: {end - start:.2f}s\n")

    def predict(self, X):
        A_final, _ = self.forward_prop(X)
        return np.argmax(A_final, axis=0)
```

### Setting the Training Parameters

```python
INPUT_DIM = X_train.shape[1]
OUTPUT_DIM = len(np.unique(y_train))
EPOCHS = 500
ALPHA = 0.05
BATCH_SIZE = 1000
layers = [INPUT_DIM, 100, 50, 50, OUTPUT_DIM]
```

### Model Training

```python
model_1 = ANN(len(layers),layers)
model_1.training(X_train, y_train, X_val, y_val, epochs=EPOCHS, batch_size=BATCH_SIZE, alpha=ALPHA)
```

### Training Cost and Accuracy Plots

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(np.array(model_1.costs), label='Training Cost')
axes[0].plot(np.array(model_1.val_costs), label='Validation Cost')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Cost')
axes[0].set_title('Training Cost')
axes[0].legend()

axes[1].plot(np.array(model_1.accuracy), label='Training Accuracy')
axes[1].plot(np.array(model_1.val_accuracy), label='Validation Accuracy')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training Accuracy')
axes[1].legend()
plt.legend()

plt.tight_layout()
plt.show()
```

### Making Predictions

```python
y_pred = model_1.predict(X_test.T)
cm = confusion_matrix(y_test, y_pred,)
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Evaluation Metrics

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average=None)
recall = recall_score(y_test, y_pred, average=None)
f1 = f1_score(y_test, y_pred, average=None)
np.set_printoptions(precision=3)

print(f"Accuracy: {acc:.3f}\n")
print("Precision:",prec)
print("Recall:   ",recall)
print("F1 Score: ",f1)
```
